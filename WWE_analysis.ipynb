```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# WWE
## World Wrestling Entertainment (WWE) is an American professional wrestling promotion.
```

```python
![WWE Logo](link-to-your-image)
```

```python
# Viewership data for WWE in millions
data = {
    "Season": [1, 2, 3, 4, 5, 6, 7, 8, 9],
    "Viewership": [4.5, 4.8, 4.2, 3.8, 3.5, 2.9, 2.8, 2.4, 1.9]
}

df = pd.DataFrame(data)
df
```

```python
# Viewership data for WWE by season
plt.figure(figsize=(10, 5))
plt.plot(df['Season'], df['Viewership'], marker='o')
plt.title('Viewership Trend of WWE by Season')
plt.xlabel('Season')
plt.ylabel('Viewership (Millions)')
plt.grid(True)
plt.show()
```

```python
# Change in viewership data for WWE by season
df['Change'] = df['Viewership'].diff()
plt.figure(figsize=(10, 5))
plt.bar(df['Season'], df['Change'], color='red')
plt.title('Change in Viewership Between Seasons of Suits')
plt.xlabel('Season')
plt.ylabel('Change in Viewership (Millions)')
plt.grid(True)
plt.show()
```

```python
## Analysis

#From the graphs, we can see a general decline in viewership from Season 1 to Season 9. The most significant drop occurred between Season 6 and Season 7, where viewership decreased by approximately **0.1 million**. This trend could be indicative of various factors, including changes in cast, plot direction, or increased competition from other shows.
```
